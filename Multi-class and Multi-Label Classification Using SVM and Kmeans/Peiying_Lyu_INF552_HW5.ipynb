### ii. SVM using Gaussian Kernel / no Standard

#### find c & gamma:

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

'''
https://jakevdp.github.io/PythonDataScienceHandbook/05.07-support-vector-machines.html
'''
#train
feature_train = train_df.iloc[:,:22]
label_train = train_df.iloc[:,-4:-1]

#test
feature_test = test_df.iloc[:,:22]
label_test = test_df.iloc[:, -4:-1]

#scaler
scaler=StandardScaler()
fea_train_scale = scaler.fit_transform(feature_train)
fea_test_scale = scaler.fit_transform(feature_test)

def from_(y_train, y_test):
    
    C = np.logspace(-3,6,num=10)
    G = 0.1*np.arange(1,21,1)
    G_range = [round(i,1) for i in G]
    tuned_parameters = [{'kernel': ['rbf'], 'gamma': G_range, 'C': C}]
    
    grid_search = GridSearchCV(SVC(), tuned_parameters, cv=10,
                                                           scoring='accuracy', refit='accuracy')
    grid_search.fit(feature_train, y_train)
    best_parameters = grid_search.best_estimator_.get_params() 
    
    y_train_pred = grid_search.predict(feature_train)
    y_test_pred = grid_search.predict(feature_test)
   
    tmp_c = 0
    tmp_g = 0
    for para, val in list(best_parameters.items()):    
        print(para, val)
        if para == 'C':
            tmp_c = val
        if para == 'gamma':
            tmp_g = val
        else:
            continue
    return( tmp_c, tmp_g)


c_ = []
g_ = []
for category in categories:
    print(category)
    y_trian = label_train[category]
    y_test = label_test[category]
    
    tmp_c, tmp_g = from_(y_trian,y_test)
    c_.append(tmp_c)
    g_.append(tmp_g)
    
    
```

# 2. K-Means Clustering on a Multi-Class and Multi-Label Data Set

### a. Use k-means clustering
#### Research:

**Hamming loss**
The fraction of the wrong labels to the total number of labels
<br>
<br>
**Hamming score**
The set of labels predicted for a sample must exactly match the corresponding set of labels in y_true.
<br>
<br>
**Hamming distance:**
Hamming distance is a metric for comparing two binary data strings. While comparing two binary strings of equal length, Hamming distance is the number of bit positions in which the two bits are different. (https://www.tutorialspoint.com/what-is-hamming-distance)
<br>
<br>
So, to calculate the distance, I compared two **2D-array (y_pred, y_true)**.  I seem 3-string-list [family, genus, species] as a unit.  Then, calculate the total number of incorrect in whole y_pred data set.  So, in some cases, the hamming distance may be similar to the hamming loss. 


## MonteCarlo Simulation 50 times

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.metrics import hamming_loss
import random
import statistics
from scipy.spatial import distance

# Loading the DATA
all_data = pd.read_csv("D:/552 Machine Learning for Data Science/homework/Homework5/Anuran Calls (MFCCs)/Frogs_MFCCs.csv")
all_data=all_data.drop('RecordID',axis=1)

cols=[e for e in all_data if e not in ( 'Family','Genus','Species')]
X=all_data.loc[:,cols]
Y=all_data.loc[:, ['Family','Genus','Species']]

total_k=[]
final_ham_dist=[]
final_ham_loss=[]
final_ham_score=[]

fam_maj_trip={p:[] for p in range(1,51)}
genus_maj_trip={p:[] for p in range(1,51)}
species_maj_trip={p:[] for p in range(1,51)}

for time in range(1,51):
    print("********         ", time,"        ********")
   
    # --------------------a. Finding the Optimal K value ------------------------
    '''
    https://stackoverflow.com/questions/51138686/how-to-use-silhouette-score-in-k-means-clustering-from-sklearn-library
    '''
    silh_avg = dict()
    for k in range(2,51):
        k_means = KMeans(n_clusters=k,init='k-means++',random_state=time).fit(X)
        labels = k_means.labels_
        silh_avg.update({k:(metrics.silhouette_score(X, labels))})
   #print("Average Silhoutte score values : ",silh_avg)
    
    optimal_k = max(silh_avg,key=silh_avg.get)
    total_k.append(optimal_k)
    K = optimal_k
    
    #-----------------------------------------------------------------------------
    tmp_X=X
    k_means_ = KMeans(n_clusters=K, random_state=time).fit(tmp_X)
    cluster_labels = k_means_.labels_
    
    clusters = pd.concat([tmp_X,Y,pd.DataFrame({'labels':cluster_labels.tolist()})],axis = 1)
    clusters['labels'].value_counts()

    
    #-----------------------  b.  determine majority labels ---------------------------
    for k in range(0,K):
        find= clusters[clusters['labels']==k]
        print('Cluster',k+1)
        print('Majority class in family - ',find['Family'].value_counts().index[0])
        print('Majority class in genus - ',find['Genus'].value_counts().index[0])
        print('Majority class in species - ',find['Species'].value_counts().index[0])

   
    #----------------------  Determine the Majority Triplet f -----------------
    '''
    http://rasbt.github.io/mlxtend/user_guide/classifier/EnsembleVoteClassifier/
    '''
    maj_trip = {k:[] for k in range(0,K)}
    for k in range(0,K):
        c_value = clusters[clusters['labels']==k]
        maj_trip[k].append(c_value['Family'].value_counts().index[0])
        maj_trip[k].append(c_value['Genus'].value_counts().index[0])
        maj_trip[k].append(c_value['Species'].value_counts().index[0])
        
        fam_maj_trip[time].append(c_value['Family'].value_counts().index[0])
        genus_maj_trip[time].append(c_value['Genus'].value_counts().index[0])
        species_maj_trip[time].append(c_value['Species'].value_counts().index[0])

    clusters['family_pred'] = 'none'
    clusters['genus_pred'] = 'none'
    clusters['species_pred'] = 'none'

    for k in range(0,K): 
        clusters['family_pred'] = np.where(clusters['labels']==k,maj_trip[k][0],clusters['family_pred'])
        clusters['genus_pred'] = np.where(clusters['labels']==k,maj_trip[k][1],clusters['genus_pred'])
        clusters['species_pred'] = np.where(clusters['labels']==k,maj_trip[k][2],clusters['species_pred'])
        

    #---------------------------   c  -----------------------------------------
    #---------------  Average hamming distance  -------------------------
    y_true = Y
    y_pred = clusters[['family_pred','genus_pred','species_pred']]
    
    tmp_mean = []
    for i in range(y_true.shape[0]):
        tmp = distance.hamming(y_true.loc[i],y_pred.loc[i])
        tmp_m= np.mean(tmp)
        tmp_mean.append(tmp_m)

    ham_dist = np.mean(tmp_mean)
    print("Hamming Distance: ", ham_dist)
    
    # ----------------   Average Hamming Score  ------------------------
    
    fam_s=hamming_loss(clusters['Family'],clusters['family_pred'])
    gen_s=hamming_loss(clusters['Genus'],clusters['genus_pred'])
    spec_s=hamming_loss(clusters['Species'],clusters['species_pred'])

    ham_loss_s=(fam_s+gen_s+spec_s)/3
    
    print("Hamming Loss : ",ham_loss_s)
    print("Hamming Score : ",1-ham_loss_s)
    print("\n")
    
    # ----------------  average & standard deviation -----------------------
    final_ham_dist.append(ham_dist)
    final_ham_loss.append(ham_loss_s)
    final_ham_score.append((1-ham_loss_s))
    
    
tmp_major=list(zip(list(fam_maj_trip.values()),
                                     list(genus_maj_trip.values()),
                                     list(species_maj_trip.values())))

major_trip={}
for i in range(0,50):
    major_trip.update({i:list(zip(tmp_major[i][0],tmp_major[i][1],tmp_major[i][2]))})

time_=range(1,51)
result=pd.DataFrame({"Iteration" : time_,
                                          "Optimal K":total_k,
                                          "Hamming Distance":final_ham_dist, 
                                          "Hamming Loss":final_ham_loss, 
                                          "Hamming Score":final_ham_score,
                                          "Majority labels (Family,Genus,Species)":list(major_trip.values())})
```

```python
print("|Average Hamming Distance|", np.mean(final_ham_dist))
print("|Standard Deviation of Hamming Distance|", np.std(final_ham_dist))

print("|Average Hamming Loss|", np.mean(final_ham_loss))
print("|Standard Deviation of Hamming Loss|", np.std(final_ham_loss))

print("|Average Hamming Score|", np.mean(final_ham_score))
print("|Standard Deviation of Hamming Score|", np.std(final_ham_score))
```
